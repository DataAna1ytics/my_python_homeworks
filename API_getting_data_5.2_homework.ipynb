## –ó–∞–¥–∞–Ω–∏–µ 3

–í—ã–≥—Ä—É–∑–∏ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `posts` –≤ —Ñ–æ—Ä–º–∞—Ç–µ –¥–∞—Ç–∞-—Ñ—Ä–µ–π–º–∞ __–≤—Å–µ –∑–∞–ø–∏—Å–∏__ —Å–æ–æ–±—â–µ—Å—Ç–≤–∞ Skyeng: https://vk.com/skyeng, 


```python
import time
```

```python
offset = 0
max_count = 100
posts = pd.DataFrame()

while True:
    url = 'https://api.vk.com/method/wall.get'
    params = {'access_token': token,
              'v': '5.89',
              'domain': 'skyeng',
              'offset': offset,
              'count': max_count
          }
    req = requests.get(url, params)
    res = req.json()
    if len(res['response']['items']) != 0:
        posts = pd.concat([posts, pd.DataFrame(res['response']['items'])])
    else:
        break
    offset = offset + max_count
    time.sleep(0.5)

```

```python
print('–ö–æ–ª-–≤–æ –∑–∞–ø–∏—Å–µ–π:', len(posts))
```
