```python
payments['payment_date'] = pd.to_datetime(payments['payment_date']) #–ø–µ—Ä–µ–≤–µ–ª –¥–∞—Ç—É –≤ datetime –¥–ª—è payments
```

```python
payments['payment_id'].groupby(payments['payment_date'].dt.year).nunique() #–≤–∏–¥–∏–º, —á—Ç–æ –≤—Å–µ 11362 –æ–ø–ª–∞—Ç—ã –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ payments –±—ã–ª–∏ –≤ 2020 –≥–æ–¥—É
```

```python
USD_RATE = 73.14 # –∫—É—Ä—Å –¥–æ–ª–ª–∞—Ä–∞

def map_currency(currency): #—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ fx_rate –º–µ—Ç–æ–¥–æ–º .apply()
    if currency == 'USD':
        return USD_RATE
    else:
        return 1

payments['fx_rate'] = payments['currency'].apply(map_currency) # –°–æ–∑–¥–∞–ª —Å—Ç–æ–ª–±–µ—Ü fx_rate
payments['amount_rub'] = payments['amount'] * payments['fx_rate'] # –°–æ–∑–¥–∞–ª —Å—Ç–æ–ª–±–µ—Ü amount_rub
payments_fx_rate = pd.DataFrame(payments) # –ü–µ—Ä–µ–∑–∞–ø–∏—Å–∞–ª –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–π df payments, –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π df –¥–ª—è —É–¥–æ–±–Ω–æ–π —Ä–∞–±–æ—Ç—ã —Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –∑–∞–¥–∞–Ω–∏—è–º–∏
```
